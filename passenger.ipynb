```python
import matplotlib.pyplot as plt

# Bar plot of survival rates by gender
df.groupby('Sex')['Survived'].mean().plot(kind='bar', color=['blue', 'pink'])
plt.title('Survival Rates by Gender')
plt.xlabel('Gender (0 = Male, 1 = Female)')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# Scatter plot: Fare vs Age
plt.scatter(df['Age'], df['Fare'], alpha=0.5, c='green')
plt.title('Fare vs Age')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
# Scatter plot for Age vs Fare
plt.scatter(df['Age'], df['Fare'], alpha=0.6, c=df['Survived'], cmap='viridis')
plt.title('Age vs Fare (Colored by Survival)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.colorbar(label='Survived')
plt.show()
```

```python
# Plot a histogram for Age distribution
df['Age'].hist(bins=20, color='orange', edgecolor='black')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot missing values count
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0]
missing_values.plot(kind='bar', color='red')
plt.title('Missing Values in Columns')
plt.xlabel('Columns')
plt.ylabel('Count')
plt.show()
```

```python
# Pie chart of passenger class distribution
class_counts = df['Pclass'].value_counts()
plt.pie(class_counts, labels=['Class 3', 'Class 1', 'Class 2'], autopct='%1.1f%%', startangle=140, colors=['gold', 'silver', 'red'])
plt.title('Passenger Class Distribution')
plt.show()
```

```python
import numpy as np

# Survival by gender and class
gender_class_survival = df.groupby(['Pclass', 'Sex'])['Survived'].mean().unstack()

# Stacked bar plot
gender_class_survival.plot(kind='bar', stacked=True, color=['blue', 'pink'])
plt.title('Survival Rate by Gender and Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.xticks(rotation=0)
plt.legend(title='Gender', labels=['Male', 'Female'])
plt.show()
```
